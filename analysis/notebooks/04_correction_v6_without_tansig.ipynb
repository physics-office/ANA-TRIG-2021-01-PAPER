```python

class fit_output:

    def __init__(self, pd_ref : float):
        self.pd_ref = pd_ref

    def __call__(self, hist : TH1 ):
        return self.calculate(hist, self.pd_ref)


    def calculate( self, hist, pd_ref):
        # Get the threshold when we not apply any linear correction
        threshold, _ = self.find_threshold( hist.ProjectionX(), pd_ref )
        # Get the efficiency without linear adjustment
        signal_eff, signal_num, signal_den = self.calculate_num_and_den_from_hist(hist, 0.0, threshold)
        # Apply the linear adjustment and fix it in case of positive slope
        slope, offset, x_dots, y_dots, errors = self.fit( hist, pd_ref )
        # Get the efficiency with linear adjustment
        corrected_eff, corrected_num, corrected_den = self.calculate_num_and_den_from_hist(hist, slope, offset)  
        return slope, offset, x_dots, y_dots, errors
    
    #
    # Find the threshold given a reference value
    #
    def find_threshold(self, th1,effref):
        nbins = th1.GetNbinsX()
        fullArea = th1.Integral(0,nbins+1)
        if fullArea == 0:
            return 0,1
        notDetected = 0.0; i = 0
        while (1. - notDetected > effref):
            cutArea = th1.Integral(0,i)
            i+=1
            prevNotDetected = notDetected
            notDetected = cutArea/fullArea
        eff = 1. - notDetected
        prevEff = 1. -prevNotDetected
        deltaEff = (eff - prevEff)
        threshold = th1.GetBinCenter(i-1)+(effref-prevEff)/deltaEff*(th1.GetBinCenter(i)-th1.GetBinCenter(i-1))
        error = 1./np.sqrt(fullArea)
        return threshold, error

    #
    # Get all points in the 2D histogram given a reference value
    #
    def get_points( self, th2 , effref):
        nbinsy = th2.GetNbinsY()
        x = list(); y = list(); errors = list()
        for by in range(nbinsy):
            xproj = th2.ProjectionX('xproj'+str(time.time()),by+1,by+1)
            discr, error = self.find_threshold(xproj,effref)
            dbin = xproj.FindBin(discr)
            x.append(discr); y.append(th2.GetYaxis().GetBinCenter(by+1))
            errors.append( error )
        return x,y,errors
    #
    # Calculate the linear fit given a 2D histogram and reference value and return the slope and offset
    #
    def fit(self, th2,effref):
        x_points, y_points, error_points = self.get_points(th2, effref )
        g = TGraphErrors( len(x_points)
                             , array.array('d',y_points,)
                             , array.array('d',x_points)
                             , array.array('d',[0.]*len(x_points))
                             , array.array('d',error_points) )
        firstBinVal = th2.GetYaxis().GetBinLowEdge(th2.GetYaxis().GetFirst())
        lastBinVal = th2.GetYaxis().GetBinLowEdge(th2.GetYaxis().GetLast()+1)
        f1 = TF1('f1','pol1',firstBinVal, lastBinVal)
        g.Fit(f1,"FRq")
        slope = f1.GetParameter(1)
        offset = f1.GetParameter(0)
        return slope, offset, x_points, y_points, error_points
    #
    # Calculate the numerator and denomitator given the 2D histogram and slope/offset parameters
    #
    def calculate_num_and_den_from_hist(self, th2, slope, offset) :
      nbinsy = th2.GetNbinsY()
      th1_num = th2.ProjectionY(th2.GetName()+'_proj'+str(time.time()),1,1)
      th1_num.Reset("ICESM")
      numerator=0; denominator=0
      # Calculate how many events passed by the threshold
      for by in range(nbinsy) :
          xproj = th2.ProjectionX('xproj'+str(time.time()),by+1,by+1)
          # Apply the correction using ax+b formula
          threshold = slope*th2.GetYaxis().GetBinCenter(by+1)+ offset
          dbin = xproj.FindBin(threshold)
          num = xproj.Integral(dbin+1,xproj.GetNbinsX()+1)
          th1_num.SetBinContent(by+1,num)
          numerator+=num
          denominator+=xproj.Integral(-1, xproj.GetNbinsX()+1)
      return numerator/denominator, numerator, denominator


def get_fit( h, pd_ref ):
    return fit_output(pd_ref)(h)
```

## Get Model:

```python
def get_model( file_name, et_bin, eta_bin, remove_last=True):
    import tensorflow as tf
    from tensorflow.keras.models import Model
    from tensorflow.keras.layers import Activation
    path=file_name.format(ET=et_bin+3,ETA=3 if eta_bin==4 else eta_bin)
    model = tf.keras.models.load_model(file_name)
    model.add(Activation('tanh'))
    new_model = Model(model.inputs, model.layers[-2].output) if remove_last else model                
    # return the models's matrix          
    return new_model

model_path = f"{data_basepath}/tunings/releases/Run2_20170505_v6/models/mc15_13TeV.361106_Zee.423300_perf_JF17.25bins.ElectronVeryLoose.Run2_v6.et{et_bin+3}_eta{eta_bin}.h5"
model = get_model(model_path, et_bin, eta_bin)
model.summary()
```

```python
model_with_tanh = get_model(model_path, et_bin, eta_bin, remove_last=False)
model_with_tanh.summary()
```

```python
def predict( data, model, col_name ):
    ring_names = [f"L2Calo_ring_{idx}" for idx in range(100)]
    rings = data[ring_names].values
    def norm1( data ):
        norms = np.abs( data.sum(axis=1) )
        norms[norms==0] = 1
        return data/norms[:,None]
    rings = norm1(rings)
    output = model.predict(rings)
    data[col_name] = output

predict( data, model, "output")
predict( data, model_with_tanh, "output_with_tanh")
data.head()
```

```python
def create_2d_hist( name, output, avgmu, xbin_size, ybin_size, ymin, ymax, xmin_percentage=1, xmax_percentage=99, xmin=None, xmax=None):
    xmin = int(np.percentile(output , xmin_percentage)) if not xmin else xmin
    xmax = int(np.percentile(output , xmax_percentage)) if not xmax else xmax
    if xmin == xmax:
        xmin = xmin - 5
        xmax = xmax + 5
    xbins = int((xmax-xmin)/xbin_size)
    ybins = int((ymax-ymin)/ybin_size)
    return rpl.hist2d.make_hist(name, output, avgmu, xbins, xmin, xmax, ybins,ymin,ymax)

def etbin2str(etbins, etidx):
    return (str(etbins[etidx]) + ' < E_{T} [GeV] < ' + str(etbins[etidx+1]) if etidx+1 < len(etbins) else
                'E_{T} > ' + ( '#infty' if etbins[-1]>9999 else str(etbins[etidx])) + ' GeV')
            
def etabin2str(etabins, etaidx):
    return (str(etabins[etaidx]) + ' < |#eta| < ' + str(etabins[etaidx+1]) if etaidx+1 < len(etabins) else
                str(etabins[etaidx]) + ' <|#eta| < 2.47')
```

```python

etidx=et_bin
etaidx=eta_bin
output = data.loc[data.target==1].output.values
avgmu = data.loc[data.target==1].avgmu.values
ymin=15
ymax=40
xmin=-5
xmax=4
xbins=0.01
ybins=1.5
ref=0.995

h = create_2d_hist("hist", output, avgmu, xbins, ybins, ymin, ymax, xmin=xmin, xmax=xmax)
slope, offset, x_dots, y_dots, errors = get_fit(h, ref)

gStyle.SetPalette(87)
fig = rpl.create_canvas("canvas", "", 500, 500)
rpl.set_figure(fig)
canvas=fig.canvas()
canvas.SetRightMargin(0.15); canvas.SetTopMargin(0.15)
canvas.SetLogz()
h.GetXaxis().SetTitle('Neural Network output (Discriminant)')
h.GetYaxis().SetTitle("<#mu>")
h.GetZaxis().SetTitle('Count'); 
rpl.add_hist(h)
h.Draw('colz')
g = TGraphErrors(len(x_dots), array.array('d',x_dots), array.array('d',y_dots), 
                     array.array('d',errors), array.array('d',[0]*len(x_dots)))
g.SetMarkerColor(kBlue)
g.SetMarkerStyle(8)
g.SetMarkerSize(1)
g.Draw("P same")
line = TLine(slope*h.GetYaxis().GetXmin()+offset,h.GetYaxis().GetXmin(), 
             slope*h.GetYaxis().GetXmax()+offset, h.GetYaxis().GetXmax())
line.SetLineColor(kBlack)
line.SetLineWidth(2)
line.Draw()
text = etbin2str(etbins, etidx)
text+= ', ' + etabin2str(etabins, etaidx)
#text+=', P_{D} = %1.2f (%1.2f) [%%]'%(pd_value*100, ref_value*100)

rpl.add_text(0.15, 0.885, text, textsize=.03)

rpl.set_atlas_label(0.15, 0.94," pp data16, #sqrt{13} TeV")
rpl.format_canvas_axes(XLabelSize=16, YLabelSize=16, XTitleOffset=0.87, ZLabelSize=16, 
                           ZTitleSize=16, YTitleOffset=0.87, ZTitleOffset=1.1)
  
fig.show()
fig.savefig("output/nn_output_versus_mu.pdf")





```
